```python
from xtb.ase.calculator import XTB
from ase.build import molecule
```

```python
mol = molecule('CH3CH2OCH3')
mol.calc = XTB()
print(mol.calc.parameters)
mol.get_potential_energy()
```

```python
mol.calc = XTB(method="GFNFF")
print(mol.calc.parameters)
mol.get_potential_energy()
```

```python
mol.calc = XTB(method="GFN1-xTB")
print(mol.calc.parameters)
mol.get_potential_energy()
```
