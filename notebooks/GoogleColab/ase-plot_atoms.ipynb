```python
from ase.build import molecule
from ase.visualize.plot import plot_atoms
```

```python

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
mol = molecule('CH3CH2NH2')
plot_atoms(mol)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5))
plot_atoms(mol, ax[0], rotation=('0x,0y,0z'))
plot_atoms(mol, ax[1], rotation=('0x,90y,0z'))
plot_atoms(mol, ax[2], rotation=('90x,0y,0z'))
```
