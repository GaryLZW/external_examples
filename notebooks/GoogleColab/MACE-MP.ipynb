```python
from mace.calculators import mace_mp
from ase.build import molecule, fcc111, add_adsorbate
mol = molecule('CO')

slab = fcc111('Cu', (5,5,6), vacuum=15)
slab.pbc=True
add_adsorbate(slab, mol, 4,  position=(3, 3))
slab.calc = mace_mp(model="medium", dispersion=False, default_dtype="float64", device='cuda')
print(slab.get_potential_energy())
```

```python
from ase.optimize import LBFGS

LBFGS(slab, trajectory='rlx.traj').run(fmax=0.01)

```
