**Global Optimisation**

Sometimes we need to overcome barriers and find multiple stable minima in a system, or find the most stable minima. To do this, we use global optimisation.

Here is a simple example for Cu2 on a Pt surface, again using the `EMT` calculator. First we load relevant modules for our work:

```python
from ase import Atoms, Atom
from ase.build import fcc110
```

With these, lets now build a relevant Pt surface and add some Cu adsorbates:

```python
# Make the Pt 110 slab.
atoms = fcc110('Pt', (2, 2, 2), vacuum=7.)

# Add the Cu2 adsorbate.
adsorbate = Atoms([Atom('Cu', atoms[7].position + (0., 0., 2.5)),
                   Atom('Cu', atoms[7].position + (0., 0., 5.0))])
atoms.extend(adsorbate)
```

`ASE` provides us with modules to make the FCC (110) surface, so I won't describe this in detail here, but do play wit the settings if you want to explore what is possible. Because we have several layers of Pt, and we just want these to stay still during the calculation (instead of moving!) we will apply some constraints so the lower layers can't move:

```python
from ase.constraints import FixAtoms, Hookean

# Constrain the surface to be fixed and a Hookean constraint between
# the adsorbate atoms.
constraints = [FixAtoms(indices=[atom.index for atom in atoms if
                                 atom.symbol=='Pt']),
               Hookean(a1=8, a2=9, rt=2.6, k=15.),
               Hookean(a1=8, a2=(0., 0., 1., -15.), k=15.),]
atoms.set_constraint(constraints)
```

Now we have a geometry model, lets actually do some energy calculations. We'll import the `EMT` calculator again and attach it to our surface model:

```python
from ase.calculators.emt import EMT

# Set the calculator.
calc = EMT()
atoms.set_calculator(calc)
```

And then run an approach for global optimisation call "Basin Hopping". This jumps between structures with minor deviations that result in "higher" energies accepted if not too big:

```python
from ase.optimize.minimahopping import MinimaHopping
# Instantiate and run the minima hopping algorithm.
hop = MinimaHopping(atoms,
                    Ediff0=2.5,
                    T0=4000.)
hop(totalsteps=10)
```

The final setting runs the calculation for 10 "hops" in this case. Once this has run we can look at the analysis of the results using some in-built functions from this module:

```python
from ase.optimize.minimahopping import MHPlot

mhplot = MHPlot()
mhplot.save_figure('summary.png')

from matplotlib import image, pyplot
# Make zoomable
#%matplotlib notebook
# Can't find a more efficient way to do this, unfortunately
imgplot = pyplot.imshow(image.imread('summary.png'))
pyplot.show()
```

The results here show multiple bits of information, but the most important is the final graph where each "flat" line indicates a minima, with a green tick showing a new structure or red arrow indicating one we've already found. The wiggles in between each minima are where the basin-hopping occurs, with many upwards energy steps allowed (hence the upwards energy paths!)
